```python
import numpy as np
```

```python
data = list(np.random.randint(10,50,100))
```

```python
print(*data, sep = '\t')
```

```python
N = len(data)
k = round(1 + 3.322 * np.log10(N))
Range = max(data) - min(data)
size = round(Range/k)
```

```python
print("Number of classes(k) : ",k)
print("Class size(c) : ",size)
```

```python
intervals = list(range(10,51,size))
#print(*intervals)

freq = []
print("_"*120)
print("|\t\t MARKS \t\t\t|\t\t NO. OF STUDENTS \t|\t PERCENTAGE FREQUENCY \t\t|")
print("_"*120)

for i in range(len(intervals)-1):
    l = intervals[i]
    h = intervals[i+1]
    # for current range l is lower limit and h is the higher limit
    
    count = 0
    for x in range(l,h): #exclusive
        count += data.count(x)
    freq.append(count)
    per = int((count/len(data))*100)
    
    print(f"|\t\t {l}-{h} \t\t\t|\t\t\t {freq[i]} \t\t|\t\t {per} \t\t\t|")
    print("_"*120)
```

#### INCLUSIVE

```python
i = min(data)
inc = []
while(i <= max(data)):
    inc.append(i)
    i += size - 1
    inc.append(i)
    i += 1
    
print(*inc)
```

```python
print("_"*110)
print("\t\t MARKS \t\t\t|\t NO. OF STUDENTS \t|\t FREQUENCY PERCENTAGE")
print("_"*110)

freq = []
j = 0
for i in range(0,len(inc) - 1,2):
    count = 0
    l = inc[i]
    h = inc[i+1]
    count = 0
    for x in range(l,h+1): #inclusive
        count += data.count(x)
    freq.append(count)
    per = int((count/len(data))*100)
    
    print(f"\t\t {l}-{h} \t\t\t|\t\t {freq[j]} \t\t|\t\t {per}")
    print("_"*110)
    
    j += 1
```
