```python
def mat2D(r,c):
    print("Enter elements for matrix: ")
    array = [[int(input()) for x in range(c)] for y in range(r)]
    return array

def add(m1,m2,r,c):
    m3 = []
    for i in range(r):
        m = []
        for j in range(c):
            m.append(m1[i][j]+m2[i][j])
        m3.append(m)
        
    print("Addition matrix: ")
    for i in range(r):
        for j in range(c):
            print(m3[i][j],end = '\t')
        print()
        
def subtract(m1,m2,r,c):
    m3 = []
    for i in range(r):
        m = []
        for j in range(c):
            m.append(m1[i][j]-m2[i][j])
        m3.append(m)
        
    print("Subtraction matrix: ")
    for i in range(len(m3)):
        for j in range(len(m3[0])):
            print(m3[i][j],end = '\t')
        print()
        
def mult_elwise(m1,m2,r,c):
    m3 = []
    for i in range(r):
        m = []
        for j in range(c):
            m.append(m1[i][j]*m2[i][j])
        m3.append(m)
        
    print("Multiplication element wise matrix: ")
    for i in range(len(m3)):
        for j in range(len(m3[0])):
            print(m3[i][j],end = '\t')
        print()

def mult(m1,m2,r1,c1,r2,c2):
    if(c1 == r2):
        m3 = []
        for i in range(r1):
            m = []
            for j in range(c2):
                m.append(0)
            m3.append(m)
            
        for i in range(len(m1)):
            for j in range(len(m2[0])):
                for k in range(len(m2)):
                    m3[i][j] += m1[i][k] * m2[k][j]
                    
        print("Matrix multiplication: ")
        for i in range(r1):
            for j in range(c2):
                print(m3[i][j],end='\t')
            print()
            
    else:
        print("Matrix multiplication not possible.")

r1 = int(input("Enter number of rows for matrix 1: "))
c1 = int(input("Enter number of columns for matrix 1: "))
m1 = mat2D(r1,c1)

r2 = int(input("Enter number of rows for matrix 2: "))
c2 = int(input("Enter number of columns for matrix 2: "))
m2 = mat2D(r2,c2)

add(m1,m2,r1,c1)
subtract(m1,m2,r1,c1)
mult_elwise(m1,m2,r1,c1)
mult(m1,m2,r1,c1,r2,c2)
```
