### <span style='color:brown'>Context</span>
The World Happiness Report is a landmark survey of the state of global happiness . The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

### <span style='color:brown'>Content
The happiness scores and rankings use data from the Gallup World Poll . The columns following the happiness score estimate the extent to which each of six factors – economic production, social support, life expectancy, freedom, absence of corruption, and generosity – contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world’s lowest national averages for each of the six factors. They have no impact on the total score reported for each country, but they do explain why some countries rank higher than others.

### Key Features in 'world happiness report':

##### <span style='color:orange'>Life Ladder: </span>
Imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you.

##### <span style='color:orange'>Log GDP per capita: </span>
At its most basic interpretation, per capita GDP shows how much economic production value can be attributed to each individual citizen. Alternatively, this translates to a measure of national wealth since GDP market value per person also readily serves as a prosperity measure.

##### <span style='color:orange'>Social support: </span>
Social support is defined in terms of social network characteristics such as assistance from family, friends, neighbours and other community members.

##### <span style='color:orange'>Healthy life expectancy at birth: </span>
Healthy life expectancy is the average life in good health - that is to say without irreversible limitation of activity in daily life or incapacities - of a fictitious generation subject to the conditions of mortality and morbidity prevailing that year

##### <span style='color:orange'>Freedom to make life choices: </span>
Freedom of choice describes an individual's opportunity and autonomy to perform an action selected from at least two available options, unconstrained by external parties.

##### <span style='color:orange'>Generosity: </span>
Generosity is the virtue of being liberal in giving, often as gifts.

##### <span style='color:orange'>Perceptions of corruption: </span>
Corruption is a form of dishonesty or criminal offense undertaken by a person or organization entrusted with a position of authority, to acquire illicit benefit

```python
q25 = happy['Ladder score'].quantile(.25)
q75 = happy['Ladder score'].quantile(.75)
mean = happy['Ladder score'].mean()

unhappiest = happy[happy['Ladder score'] < q25]
middle = happy[(happy['Ladder score'] > q25) & (happy['Ladder score'] < q75)]
happiest = happy[happy['Ladder score'] > q75]

fig,ax = plt.subplots(nrows=1, ncols=3, figsize=(20,6))
ax = plt.subplot(1, 3, 1).set(title='unhappy', ylim = [0, 25])
sns.countplot(x='Regional indicator', data=unhappiest)
plt.xticks(rotation=45, ha='right')
ax = plt.subplot(1, 3, 2).set(title='middle', ylim = [0, 25])
sns.countplot(x='Regional indicator', data=middle)
plt.xticks(rotation=45, ha='right')
ax = plt.subplot(1, 3, 3).set(title='happy', ylim = [0, 25])
sns.countplot(x='Regional indicator', data=happiest)
plt.xticks(rotation=45, ha='right')
```

```python
happiest_chart = happiest.groupby('Regional indicator', as_index=False).agg({'Country name': 'count'})
for region in (happiest_chart.sort_values(by='Country name', ascending=False)['Regional indicator']):
    print()
    print('The happiest countries in ' + region + ' are:', end=' ')
    happy_list = list(happiest[happiest['Regional indicator'] == region]['Country name'])
    print(*happy_list, sep=', ')
    print('-'*120)
```

```python
unhappiest_chart = unhappiest.groupby('Regional indicator', as_index=False).agg({'Country name': 'count'})
for region in (unhappiest_chart.sort_values(by='Country name', ascending=False)['Regional indicator']):
    print()
    print('The unhappiest countries in ' + region + ' are:', end=' ')
    unhappy_list = list(unhappiest[unhappiest['Regional indicator'] == region]['Country name'])
    print(*unhappy_list, sep=', ')
    print('-'*120)
```

```python
regions = happy['Regional indicator'].unique()

for region in regions:
    d1 = happy[happy['Regional indicator'] == region][['Country name', 'Regional indicator', 'Ladder score']]
    upper = d1[d1['Ladder score'] == d1['Ladder score'].max()].values
    lower = d1[d1['Ladder score'] == d1['Ladder score'].min()].values      
    print('Happiest in ' + region + ' is ' + upper[0][0] + ' with score ' + str(upper[0][2]))
    print('Unhappiest in ' + region + ' is ' + lower[0][0] + ' with score ' + str(lower[0][2]))
    print(end='---------------------------------------------------------------')
    print()
```

```python
df= pd.read_csv("world-happiness-report.csv")
```

```python
df.fillna(0,inplace= True)
```

```python
#creating total overall score column
df["Total_score"]=(df["Life Ladder"]+df["Log GDP per capita"]+df["Social support"]+df["Healthy life expectancy at birth"]+df["Freedom to make life choices"]+df["Generosity"]+df["Positive affect"])- (df["Perceptions of corruption"]+df["Negative affect"])
```

```python
df.head()
```

```python
#Top 20 happiest countries based on overall total score/index
#Singapore,Switzerland,Norway can be considered most happy countries based on overall total scores
#Other notable countries in the top 20 are Australia,United Kingdom,Japan,Spain,Canada,etc
x= df.groupby("Country name")["Total_score"].mean().sort_values(ascending= False).head(20)
sns.set_style("whitegrid")
plt.figure(figsize=(6,9))
sns.barplot(x.values,x.index)
plt.show()
```

```python
#Least 20 happiest countries based on overall total score
#Somaliland region,Kosovo,North Cyprus have the lowest happiness score 
#Other notable countries having lowest happiness index are Zimbabwe,Haiti,Nigeria,Ivory Coast,etc
x= df.groupby("Country name")["Total_score"].mean().sort_values(ascending= False).tail(20)
sns.set_style("whitegrid")
plt.figure(figsize=(6,9))
sns.barplot(x.values,x.index)
plt.show()
```

```python
#Correlation between GDP Per capita and the total overall happiness index score
#Majority of the total scores are directly propotional to the GDP
#The gdp per capita plays a vital role in determining the overall happiness index
#Total overall happiness index grows as gdp grows for a country
plt.figure(figsize=(7,15))
sns.set_theme(style="whitegrid")
ax= sns.jointplot(x="Log GDP per capita",y= "Total_score",kind= "hex",data= df,height= 9)
plt.show()
```

```python
#Correlation between Life expectancy at birth and the total overall index
#The below plot shows similar relation to that of GDP with overall score
#Life expectancy at birth score of a country can pretty much show how the overall hapiness of a country score is
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
sns.jointplot(x="Total_score",y= "Healthy life expectancy at birth",data= df,height=9 )
plt.show()
```

```python
#How social support impacts overall happiness score
#Countries having high score for total happiness index correlate with the high score of social support

sns.set_theme(style="darkgrid")
sns.relplot(x="Total_score",y= "Social support",data= df,size= "Total_score",height= 9)
```

```python
#Total happiness score corelation with the life ladder score of the countries
sns.set_theme(style="whitegrid")
sns.jointplot(x="Life Ladder",y= "Total_score",data= df,kind= "reg",height= 9)
```

```python
#GDP per capita and perceptions of corruption correlation
#Countries with higher GDP per capita do indeed have a higher perceptions of corruptions
sns.set_theme(style="darkgrid")
sns.jointplot(x="Log GDP per capita",y= "Perceptions of corruption",data= df,kind= "hex",height= 9)
```

```python
#Positive affects score correlation with freedom to make life choices
plt.figure(figsize=(6,6))
sns.jointplot(x="Freedom to make life choices",y= "Positive affect",data= df ,hue="year",height= 9)
```
