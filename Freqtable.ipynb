```python
import numpy as np
import random
```

```python
data = np.random.randint(0,51,200)
print("Ungrouped data:\n")
print(*data,sep = '\t')
```

```python
sort = sorted(data)
```

```python
freq = dict()
for i in sort:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
        
print("\tMarks","No. of students",sep = '\t\t\t|\t\t\t')
print('_'*80)
        
for k,v in freq.items():
    print(f'\t{k}',v,sep = '\t\t\t|\t\t\t')
    print('_'*80)
    
print("\tTotal",len(data),sep = '\t\t\t|\t\t\t')
```

```python
N = len(data)

K = 1 + 3.322 * np.log10(N)
K = int(np.ceil(K))

Range = 50
c = Range/K
c = int(np.ceil(c))

print(f"Number of classes: {K}\nClass width: {c}\n")

print("\tMarks","No. of students",sep = '\t\t\t|\t\t\t')
print('_'*80)

i = 0
total = 0
final = 0

while(K):
    total = 0
    l = i
    r = i+c
    print(f'\t{l}-{r}',end = '\t\t\t|\t\t\t')
    C = c
    
    while(C):
        if i in freq:
            total += freq[i]
        C -= 1
        i += 1
    
    final += total
    print(total)
    print('_'*80)
    K -= 1
    
print("\tTotal",final,sep = '\t\t\t|\t\t\t')
```
